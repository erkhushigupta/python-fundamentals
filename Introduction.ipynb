String slicing

```python
text = "Hello, World!"
sliced_text = text[7:12]
print(sliced_text)  # Output: World
```

From Beginning

```python
text = "Hello, World!"
sliced_text = text[:5]
print(sliced_text)  # Output: Hello
```

Until the end

```python
text = "Hello, World!"
sliced_text = text[7:]
print(sliced_text)  # Output: World!

```

Slicing with a step value:

```python
text = "Hello, World!"
sliced_text = text[::2]
print(sliced_text)  # Output: Hlo ol!
```

Negative slicing from the end of the string:

```python
text = "Hello, World!"
sliced_text = text[-6:-1]
print(sliced_text)  # Output: World
```

Reverse a string using slicing

```python
text = "Hello, World!"
reversed_text = text[::-1]
print(reversed_text)  # Output: !dlroW ,olleH
```

LIST COMPREHENSION


```python
# Creating a new list by iterating over an existing list

numbers = [1, 2, 3, 4, 5]
squared_numbers = [num ** 2 for num in numbers]
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

```python
# Filtering elements from a list based on a condition

numbers = [1, 2, 3, 4, 5]
even_numbers = [num for num in numbers if num % 2 == 0]
print(even_numbers)  # Output: [2, 4]
```

```python
# Creating a new list by manipulating elements from an existing list

names = ["Alice", "Bob", "Charlie"]
uppercase_names = [name.upper() for name in names]
print(uppercase_names)  # Output: ['ALICE', 'BOB', 'CHARLIE']
```

```python
# Creating a list of tuples from two separate lists

numbers = [1, 2, 3]
letters = ['A', 'B', 'C']
combined_list = [(num, letter) for num in numbers for letter in letters]
print(combined_list)  # Output: [(1, 'A'), (1, 'B'), (1, 'C'), (2, 'A'), (2, 'B'), (2, 'C'), (3, 'A'), (3, 'B'), (3, 'C')]
```

```python
# Conditional expression in list comprehension

numbers = [1, 2, 3, 4, 5]
new_numbers = [num if num % 2 == 0 else num * 2 for num in numbers]
print(new_numbers)  # Output: [2, 2, 6, 4, 10]
```

```python
# List comprehension with nested loops
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened_matrix = [num for row in matrix for num in row]
print(flattened_matrix)  # Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

Dictionary as a Histogram

```python
text = "Lorem ipsum dolor sit amet consectetur adipiscing elit Lorem ipsum"
words = text.split()
count = {}
for word in words:
    if word in count:
        count[word] += 1
    else:
        count[word] = 1

print(count)
# Output: {'Lorem': 2, 'ipsum': 2, 'dolor': 1, 'sit': 1, 'amet': 1, 'consectetur': 1, 'adipiscing': 1, 'elit': 1}
```

Set operations in Python

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Union of two sets
union_set = set1.union(set2)
print(union_set)  # Output: {1, 2, 3, 4, 5, 6, 7, 8}

# Intersection of two sets
intersection_set = set1.intersection(set2)
print(intersection_set)  # Output: {4, 5}

# Difference of two sets
difference_set = set1.difference(set2)
print(difference_set)  # Output: {1, 2, 3}

# Symmetric difference of two sets
symmetric_difference_set = set1.symmetric_difference(set2)
print(symmetric_difference_set)  # Output: {1, 2, 3, 6, 7, 8}
```
